For the given dataset that contains immigration details to Canada from 1980 to 2013,

• Create an area plot for top 6 immigrant countries from 1990 to 2013

• Create and year-wise immigrant bar chart from India to Canada during the period of 1980 to 2013.

• Create a boxplot for Indian, Phillipin and China immigrants.

• Show the total no. of immigrants from India and France countries using Area Chart and Pie chart.

• Create a scatter Histogram for the immigrants from Fiji and Singapore in the year 2013

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("canadian_immegration_data.csv")
df.set_index("Country", inplace=True)
```

Area plot

```python
data = df.loc[['India','China','Philippines','Pakistan','France','Bangladesh'],
              '1990':'2013'].T
data.plot(kind='area')
plt.title("Top 6 Immigrant Countries (1990–2013)")
plt.xlabel("Year")
plt.ylabel("Immigrants")
plt.show()
```

Bar chart

```python
years = list(range(1980, 2014))
immigrants = df.loc['India', '1980':'2013']
plt.bar(years, immigrants)
plt.title("Year-wise Immigration from India to Canada")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.show()

```

Boxplot

```python
india = df.loc['India', '1980':'2013']
philippines = df.loc['Philippines', '1980':'2013']
china = df.loc['China', '1980':'2013']
plt.boxplot([india, philippines, china])
plt.title("Boxplot of Immigration Data")
plt.ylabel("Number of Immigrants")
plt.show()
```

Area chart -India and France

```python
years = list(range(1980, 2014))

india = list(df.loc['India', '1980':'2013'])
france = list(df.loc['France', '1980':'2013'])
plt.fill_between(years, india, label='India')
plt.fill_between(years, france, label='France')
plt.title("Immigration from India and France")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.legend()
plt.show()

```

Pie chart

```python
india_total = sum(df.loc['India', '1980':'2013'])
france_total = sum(df.loc['France', '1980':'2013'])
values = [india_total, france_total]
labels = ['India', 'France']
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Total Immigration Share")
plt.show()
```

Scatter and  Histogram

```python
values = df.loc[['Fiji','Singapore'],'2013']
plt.scatter(values.index, values.values)
plt.title("Immigration in 2013")
plt.show()
plt.hist(values.values)
plt.title("Histogram (2013)")
plt.show()
```
